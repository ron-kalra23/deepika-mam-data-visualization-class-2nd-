```python
sns.distplot(Tips['tip'])
```

```python
# in distribution plot automatically bar graph and line is coming 
```

```python
# This line is known as KDE = kernel density estimation 
```

```python
# by default KDE is true so everytime in distplot this line will appear 
```

```python
sns.distplot(Tips['total_bill'] , bins = 50 , kde = False ) # now kde is not appearing 
# this means no curve required 
```

```python
# There is also a kde plot 
sns.kdeplot(Tips['tip'])
```

```python
sns.kdeplot(Tips['total_bill'])
```

Note --> kde plot will not give default bars that will come in distribution plot . 

composite view onto multivariate dataset 

```python
sns.jointplot(data = Tips , x = 'total_bill' , y = 'tip')
```

```python
# we can see above that with increasing in total bill there is a increase in tips. 
# but total_bill more than 30 are less 
# in joint plot we can take multiple plot like scatter and bar plot 
```

the bar above total_bill is showing for total_bill only and the bar above the tip showing for tips only .

```python
#sns .rugplot?
sns.rugplot(data = Tips , x = 'tip')
```
